```python
# CASE文（条件式）の練習用ファイル
import pandas as pd
import sqlite3
```

```python
class Sql:
    def __init__(self, df:pd.DataFrame, conn:sqlite3.Connection):
        self.df = df
        self._conn = conn
    
    
    def conversion_sql(self, name:str) -> None:
        data = self.df.to_sql(
            name= name,
            con=self._conn,
            if_exists="replace",
            index=False
        )
        return data
    
    
    def result_sql(self, query:str) -> None:
        result = pd.read_sql(
            query,
            con=self._conn
        )
        return print(result)
```

```python
df = pd.DataFrame({
    "id": [1, 2, 3, 4, 5, 6],
    "name": ["Aさん", "Bさん", "Cさん", "Dさん", "Eさん", "Fさん"],
    "points": [200, 100, 500, 300, 400, 600]
})

print(df)
```

```python
conn = sqlite3.connect(":memory:")
sql = Sql(df=df, conn=conn)
sql.conversion_sql("game")
```

```python
# 条件式
sql.result_sql(
    """SELECT name, points,
        CASE
            WHEN points >= 500 THEN '上級'
            WHEN points >= 400 THEN '中級'
            ELSE '初級'
        END AS rank
        FROM game
    """
)
```

```python
# 条件式＆並び替え（昇順）
sql.result_sql(
    """SELECT name, points,
        CASE
            WHEN points >= 500 THEN '上級'
            WHEN points >= 400 THEN '中級'
            ELSE '初級'
        END AS rank
        FROM game
        ORDER BY points ASC
    """
)
```
