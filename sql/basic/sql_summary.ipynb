```python
# SQL集計関数の練習用ファイル
import pandas as pd
import sqlite3
```

```python
class SQL:
    def __init__(self, df:pd.DataFrame):
        self._conn = sqlite3.connect(":memory:")
        self.df = df
        df["date"] = df["date"].dt.strftime("%Y-%m-%d")
        
    def conversion_sql(self, name:str) -> None:
        df = self.df.to_sql(
            name= name,
            con=self._conn,
            if_exists="replace",
            index=False
        )
        return df
    
    def result_sql(self, query:str) -> None:
        result = pd.read_sql(
            query,
            con=self._conn
        )
        return print(result)
```

```python
df = pd.DataFrame({
    "id": [1, 2, 3, 4, 5, 6, 7, 8],
    "name": ["Aさん", "Bさん", "Cさん", "Dさん", "Eさん", "Fさん", "Gさん", "Hさん"],
    "department": ["経理", "経理", "営業", "営業", "総務", "総務", "開発", "開発"],
    "salary": [40000, 20000, 80000, 50000, 60000, 40000, 50000, 90000],
    "date" : pd.to_datetime(
        ["2026-5-10", "2026-6-30", "1987-3-12", "2003-8-10", "2014-2-1", "2020-5-29", "2025-4-1", "2010-6-10"]
    )
})
print(df)
```

```python
sql = SQL(df=df)
sql.conversion_sql(name="summary")
```

```python
# 合計
sql.result_sql(
    query=
    """SELECT SUM(salary) AS total_salary
        FROM summary
    """
)
```

```python
# 最大
sql.result_sql(
    query= 
    """SELECT department, MAX(salary) AS max_salary
        FROM summary
    """    
)
```

```python
# 最小
sql.result_sql(
    query=
    """SELECT department, MIN(salary) AS min_salary
        FROM summary
    """
)
```

```python
sql.result_sql(
    query=
    """SELECT AVG(salary) AS average_salary
        FROM summary
    """
)
```

```python
sql.result_sql(
    query=
    """SELECT COUNT(id) AS total_officer
        FROM summary
    """
)
```
