```python
# SQL基本文法の練習用ファイル
import pandas as pd
import sqlite3
```

```python
df = pd.DataFrame({
    "id": [1, 2, 3, 4, 5, 6],
    "name": ["Aさん", "Bさん", "Cさん", "Dさん", "Eさん", "Fさん"],
    "department": ["経理", "経理", "営業", "営業", "総務", "総務"],
    "salary": [50000, 50000, 70000, 70000, 80000, 80000],
    "date" : pd.to_datetime(
        ["2026-5-10", "2026-6-30", "1987-3-12", "2003-8-10", "2014-2-1", "2020-5-29"]
    )
})
df["date"] = df["date"].dt.strftime("%Y-%m-%d")
print(df)
```

```python
# 取得・追加・更新・削除のSQLクエリ
conn =sqlite3.connect(":memory:")

df.to_sql(
    name="basic",
    con=conn,
    if_exists="replace",
    index=False
)
```

```python
# 追加
conn.execute(
    """INSERT INTO basic
        VALUES ('7', 'Gさん', '設計', '50000', '2014-09-20')
    """
)

conn.commit()

result = pd.read_sql(
    """SELECT * FROM basic
    """,
    con=conn,
)

print(result)
```

```python
# 更新
conn.execute(
    """UPDATE basic
        SET salary = 60000
        WHERE department = '経理'
    """
)

conn.commit()

result = pd.read_sql(
    "SELECT * FROM basic",
    con=conn,
)

print(result)
```

```python
# 削除
conn.execute(
    """DELETE FROM basic
        WHERE name = 'Cさん'
    """
)

conn.commit()

result = pd.read_sql(
    "SELECT * FROM basic",
    con=conn,
)

print(result)
```

```python
# 並び替え
conn.execute(
    """SELECT * FROM basic
        ORDER BY salary DESC
    """
)

conn.commit()

result = pd.read_sql(
    "SELECT * FROM basic",
    con=conn,
)

print(result)
```

```python
# グループ化
result = pd.read_sql(
    """SELECT department, SUM(salary) AS total_salary
        FROM basic
        GROUP BY department
    """,
    con=conn,
)

print(result)
```

```python
# グループ後の条件指定
result = pd.read_sql(
    """SELECT department, SUM(salary) AS total_salary
        FROM basic
        GROUP BY department
        HAVING SUM(salary) >= 100000
    """,
    con=conn,
)

print(result)
```

```python
# 結果の並べ替え→先頭から数行だけ取得
result = pd.read_sql(
    """SELECT *
        FROM basic
        ORDER BY name ASC LIMIT 4 OFFSET 1
    """,
    con=conn,
)

print(result)
```
