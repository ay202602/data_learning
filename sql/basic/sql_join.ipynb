```python
# JOIN（結合）のクエリに関する練習用ファイル
import pandas as pd
import sqlite3
```

```python
class Sql:
    def __init__(self, df:pd.DataFrame, conn:sqlite3.Connection):
        self.df = df
        self._conn = conn
        
    def conversion_sql(self, name:str,) -> None:
        data = self.df.to_sql(
            name= name,
            con=self._conn,
            if_exists="replace",
            index=False
        )
        return data
    
    def result_sql(self, query:str) -> None:
        result = pd.read_sql(
            query,
            con=self._conn
        )
        return print(result)
```

```python
df_customers = pd.DataFrame({
    "customer_id": [1, 2, 3, 4],
    "name": ["Aさん", "Bさん", "Cさん", "Dさん"],
    "city": ["東京", "大阪", "福岡", "東京"]
})

df_orders = pd.DataFrame({
    "order_id": [101, 102, 103, 104],
    "customer_id": [1, 1, 2, 99],
    "product": ["ペン", "ノート", "消しゴム", "定規"],
    "amount": [100, 200, 50, 150]
})

print(df_customers)
print(f"\n{df_orders}")
```

```python
conn = sqlite3.connect(":memory:")

sql_customers = Sql(df_customers, conn=conn)
sql_customers.conversion_sql("customers")

sql_orders = Sql(df_orders, conn=conn)
sql_orders.conversion_sql("orders")
```

```python
# 内部結合(条件が合わない行は結果から消滅)
sql_customers.result_sql(
    """SELECT *
        FROM customers
        INNER JOIN orders
        ON customers.customer_id = orders.customer_id
    """
)
```

```python
# 左外部結合（左の行を強制的にすべて表示）（条件に合わないものは削除されず表示される）
sql_customers.result_sql(
    """SELECT *
        FROM customers
        LEFT JOIN orders
        ON customers.customer_id = orders.customer_id
    """
)
```

```python
# 右外部結合（右の行を強制的にすべて表示）（条件に合わないものは削除されず表示される）
sql_customers.result_sql(
    """SELECT *
        FROM customers
        RIGHT JOIN orders
        ON customers.customer_id = orders.customer_id
    """
)
```
